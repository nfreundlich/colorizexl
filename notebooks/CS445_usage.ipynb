```python
%matplotlib inline
import os
import sys
import matplotlib.pyplot as plt
sys.path.insert(0,'../')
```

```python
# import ColorizeXL module
from colorizexl import ColorizeXL
```

```python
# Colorize a grayscale image
input_grayscale_name = '../samples/boy.png'
input_annotation_name = '../samples/boy_annotated.png'
colorizer = ColorizeXL(
        input_grayscale_name, 
        input_annotation_name, 
        recolorize=False
    )

output = colorizer.colorize(step_size=100, 
                            overlap=2, 
                            n=10)
plt.imshow(output)
```

```python
# Reolorize a color image
input_original_name = '../samples/chair.png'
input_annotation_name = '../samples/chair_annotated.png'
colorizer = ColorizeXL(
        input_original_name, 
        input_annotation_name, 
        recolorize=True
    )

output_recolorize = colorizer.colorize(step_size=100, 
                                       overlap=2, 
                                       n=10)
plt.imshow(output_recolorize)
```
